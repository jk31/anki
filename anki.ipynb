```python
import pandas as pd
import genanki
```

```python
df = pd.read_csv('data/translations_june.csv', header=None)
```

```python
df[20:40]
```

```python
for index, row in df.iterrows():
    if row[0] == "Deutsch":
        german_word = row[2]
        englisch_word = row[3]
        row[0] = "Englisch"
        row[1] = "Deutsch"
        row[2] = englisch_word
        row[3] = german_word
```

```python
my_model = genanki.Model(
  12345,
  'eng-ger',
  fields=[
    {'name': 'eng'},
    {'name': 'ger'},
  ],
  templates=[
    {
      'name': 'Card 1',
      'qfmt': '{{ger}}',
      'afmt': '{{FrontSide}}<hr id="answer">{{eng}}',
    },
  ])
```

```python
my_deck = genanki.Deck(
  54321,
  'eng-ger')
```

```python
for index, row in df.iterrows():
    
    new_note = genanki.Note(
        model=my_model,
        fields=[row[2], row[3]])
    
    my_deck.add_note(new_note)
```

```python
genanki.Package(my_deck).write_to_file('eng_ger.apkg')
```
